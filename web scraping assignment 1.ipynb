```python
#Ans3: program to display IMDB's top rated indian 100 movies data i.e name, IMDB rating, year of release
# importing the required libraries
from bs4 import BeautifulSoup
import requests

page = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page
```

```python
Soup = BeautifulSoup(page.content)
Soup
```

```python
name= Soup.find_all(class_="titleColumn")
name
```

```python
movie_name = []
for i in name:
    for j in i.find_all("a"):
        movie_name.append(j.text.replace("\n",""))
movie_name
```

```python
year = Soup.find_all("span", class_="secondaryInfo")
year
```

```python
Year = []
for i in year:
    Year.append(i.text)
Year
```

```python
rating = Soup.find_all( class_="ratingColumn imdbRating")
rating
```

```python
Rating = []
for i in rating:
    Rating.append(i.text.replace("\n",""))
Rating
```

```python
import pandas as pd
```

```python
web_series=pd.DataFrame({})
web_series['movie_name']=movie_name[:100]
web_series['Year']=Year[:100]
web_series['Rating']=Rating[:100]
web_series
```

```python
#ans:4 program for find book name, author, review
from bs4 import BeautifulSoup
import requests

page = requests.get("https://bookpage.com/reviews")
page
```

```python
Soup = BeautifulSoup(page.content)
Soup
```

```python
name= Soup.find_all(class_="flex-article-content")
name
```

```python
book_name = []
for i in name:
    for j in i.find_all("a"):
        book_name.append(j.text.replace("\n",""))
book_name
```

```python
author = Soup.find_all( class_="sans bold")
author
```

```python
Author = []
for i in author:
    Author.append(i.text.replace("\n",""))
Author
```

```python
review = Soup.find_all( class_="excerpt")
review
```

```python
Review = []
for i in review:
    Review.append(i.text.replace("\n",""))
Review
```

```python
import pandas as pd
```

```python
book_series=pd.DataFrame({})
book_series['book_name']=book_name[:5]
book_series['Author']=Author[:5]
book_series['Review']=Review[:5]
book_series
```
